```python
#importing dependencies
import pandas as pd
import ast
import zipfile
from sqlalchemy import create_engine
```

```python
#reading and displaying in the csv that holds our actors information
actors_file = "Resources/credits.csv"
actors_df = pd.read_csv(actors_file)
actors_df.head()
```

```python
#only keeping the ID and cast columns
actors_df=actors_df[["id","cast"]]
actors_df.head()
```

```python
#dropping any row with empty cells
actors_df.dropna(how="any",inplace=True)
```

```python
#resetting index in case any rows were dropped
actors_df.reset_index
actors_df.head()
```

```python
#unpacking the cast column by reading into each dictionary and pulling out the actor name
character_name = []
for index,row in actors_df.iterrows():
    id = row["id"]
    row = ast.literal_eval(row["cast"])
    for i in range(len(row)):
        name=row[i]["name"]
        character_name.append({"name":name,"id":id})
```

```python
actors_df = pd.DataFrame(character_name)
actors_df.set_index('id').head()
```

```python
actors_df = actors_df.rename(columns = {"id":"Movie_ID","name":"Actors"}).set_index('Movie_ID')
```

```python
#Outputting the dataframe to a csv file
actors_df.to_csv("Output/Actors.csv", index=True, header=True)
```

```python
#Reading and displaying in the csv that holds our rating information
rating_file = 'Resources/ratings.csv' 
rating_df = pd.read_csv(rating_file) 
rating_df.set_index('movieId').head()
```

```python
rating_df = rating_df[["movieId","rating"]]
```

```python
rating_df.rename(columns = {"movieId":"Movie_ID","rating":"Rating"}).set_index('Movie_ID').head()
```

```python
#Outputting the dataframe to a csv file
rating_df.to_csv("Output/Rating.csv", index=True, header=True)
```

```python

genres_file = "Resources/movies_metadata.csv"
genres_df = pd.read_csv(genres_file)
genres_df = genres_df[~genres_df["id"].str.contains("-")]
genres_df.set_index('id').head()
```

```python
genres_df["genres"][0][0]
type(ast.literal_eval(genres_df["genres"][0])[0])
```

```python
genre = []
for index,row in genres_df.iterrows():
    id = row["id"]
    row = ast.literal_eval(row["genres"])
    for i in range(len(row)):
        name=row[i]["name"]
        genre.append({"name":name,"id":id})
```

```python
genres_df = pd.DataFrame(genre)
```

```python
genres_df.set_index('id').head()
```

```python
genres_df.rename(columns = {"id":"Movie_ID","name":"Genre"}).set_index('Movie_ID').head()
```

```python
genres_df.to_csv("Output/Genres.csv", index=True, header=True)
```
